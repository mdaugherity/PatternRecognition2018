```python
plt.plot([1,2,4,7]) #this is a nice plot of four points
plt.show()
```

```python
#now let's plot points 
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 10, 13]
plt.axis([0, 6, 2, 14])
plt.plot(x,y,'ro') # the 'r' says what color / the '0' says to do circle markers
plt.show()
```

```python
x = np.arange(1,19,.4)
y1 = np.log10(x)
y2 = 0.01*x**2
y3 = 0.9*np.sin(x)
plt.plot(x, y1, 'r-', label = 'y1')
plt.plot(x, y2, 'b^', label = 'y2')
plt.plot(x, y1, 'go', label = 'y3')
plt.plot(x, y1+y2+y3, '+', label = 'y1 + y2 + y3')
plt.legend(loc=2) # put a legend somewhere on the screen
plt.show()
```

```python
# here's an example of how to do subplots
def func(x):
    return np.sin(2*np.pi*x)
x1 = np.arange(0, 4, 0.1)
x2 = np.arange(0, 4, 0.1)
y1 = func(x1)
y2 = func(x2)
yln = y1 + .1*np.random.randn(len(x1)) 

plt.figure()

plt.subplot(211) # this makes two subplots, this is the top one

plt.plot(x1, y1, 'bo', x2, y2, 'r:')

plt.subplot(212) # this is the bottom subplot
plt.plot(x1, yln, 'bo', x2, y2, 'r:')

plt.show()
```
