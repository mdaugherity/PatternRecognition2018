```python
Xcircle, ycircle = datasets.make_circles()

num_blobs = 20
Xblobs,yblobs = datasets.make_blobs(n_samples=10*num_blobs,centers=num_blobs,random_state=42)
yblobs = np.floor_divide(yblobs,num_blobs/2) # changing from 6 classes to 2

Xmoons, ymoons = datasets.make_moons()

Xiris,yiris = datasets.load_iris(return_X_y=True)
Xiris = Xiris[yiris!=2,1:3] # dropping 3rd class and last 2 features
yiris = yiris[yiris!=2] 

Xiris,yiris = datasets.make_blobs(n_samples=100,centers=2,random_state=42,cluster_std=4)
                                 
# listify 
Xdata = (Xiris,Xblobs,Xmoons,Xcircle)
ydata = (yiris,yblobs,ymoons,ycircle)
```

```python
# STOLEN FROM XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
plt.figure(figsize=(16,4))

for i,(X,y) in enumerate(zip(Xdata,ydata)):
    plt.subplot(1,4,i+1)
    X0 = X[:,0]
    X1 = X[:,1]
    plt.scatter(X0, X1,c=y,cmap='coolwarm')
    
    
plt.show()
```

```python
plt.figure(figsize=(16,4))

ind = 1

X=Xdata[ind]
X0 = X[:,0]
X1 = X[:,1]
xx, yy = make_meshgrid(X0, X1)
y = ydata[ind]


ax = plt.subplot(1,4,1)
clf = neural_network.MLPClassifier(hidden_layer_sizes=(2),max_iter=10000)
clf.fit(X,y)
plot_contours(ax,clf,xx,yy,cmap='coolwarm',alpha=0.4)
plt.scatter(X0, X1, c=y,cmap='coolwarm')

ax = plt.subplot(1,4,2)
clf = neural_network.MLPClassifier(hidden_layer_sizes=(20),max_iter=10000)
clf.fit(X,y)
plot_contours(ax,clf,xx,yy,cmap='coolwarm',alpha=0.4)
plt.scatter(X0, X1, c=y,cmap='coolwarm')

ax = plt.subplot(1,4,3)
clf = neural_network.MLPClassifier(hidden_layer_sizes=(10,4,2,2),max_iter=10000)
clf.fit(X,y)
plot_contours(ax,clf,xx,yy,cmap='coolwarm',alpha=0.4)
plt.scatter(X0, X1, c=y,cmap='coolwarm')


ax = plt.subplot(1,4,4)
clf = neural_network.MLPClassifier(hidden_layer_sizes=(20,20),max_iter=10000)
clf.fit(X,y)
plot_contours(ax,clf,xx,yy,cmap='coolwarm',alpha=0.4)
plt.scatter(X0, X1, c=y,cmap='coolwarm')




plt.show()
```

```python
#hiddens = ((2),(4),(6),(20),(10,4,2),(20,20))

hiddens = ((2),(4),(6),(8),(10),(10,10),(2,2,2),(2,2))

plt.figure(figsize=(16,16))


for i,(X,y) in enumerate(zip(Xdata,ydata)):
    X0 = X[:,0]
    X1 = X[:,1]
    xx, yy = make_meshgrid(X0, X1)
    
    for j,hidden in enumerate(hiddens):
        ax = plt.subplot(len(hiddens),len(Xdata),i+1+len(Xdata)*j)
        ax.set_title(hidden)
        ax.set_xticks(())
        ax.set_yticks(())
        clf = neural_network.MLPClassifier(hidden_layer_sizes=hidden,max_iter=10000,random_state=42)
        clf.fit(X,y)
        plot_contours(ax,clf,xx,yy,cmap='coolwarm',alpha=0.4)
        plt.scatter(X0, X1, c=y,cmap='coolwarm')

plt.tight_layout()
plt.show()
```
