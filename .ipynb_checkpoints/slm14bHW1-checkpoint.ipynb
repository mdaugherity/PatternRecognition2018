## Problem Description
 1) Determine the overall "best" classifier. 2) Also discuss which classifier is best maximizing precision, and which also minizes false positives. 3)Find the optimal probability threshhold for your best classifier and use it to classify the points in X_new.


## Solution Method
The first step will be to create the ROC plots of each of the classifiers. I will use the code that we used earlier in the week to plot the data as we send the test datasets through each classifier. Then using the scores retreived through the different methods I will use sklearn.metrics to create the ROC plots and analyze the efficacy of each of the classifying mechanisms. The ROC plots will give us a method of visualizing the optimal threshold values for each classifier and we can calculate the area under the ROC curve.

```python
## Input
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics


X, y = datasets.make_classification(n_samples=100, n_features=8, n_informative=3, 
                                      n_classes=2, class_sep=0.7, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

X_new, _ = datasets.make_classification(n_samples=5, n_features=8, n_informative=3, 
                                      n_classes=2, class_sep=0.7, random_state=42)

from sklearn.svm import SVC
from sklearn import tree
from sklearn import neural_network

svc = SVC(kernel='poly', C=1,probability=True)
dtr = tree.DecisionTreeClassifier(max_depth=4)
ann = neural_network.MLPClassifier(random_state=1)


clf_list = (svc,dtr,ann) # list o' classifiers

# Train all the classifiers in a for loop, just because we can
for clf in clf_list:
    clf.fit(X_train,y_train)
```

```python
## Analysis
from sklearn import metrics

#Using Train Data Set
y_scoresvc = svc.predict_proba(X_train)[:,1]
fprsvc, tprsvc, threshsvc = metrics.roc_curve(y_train,y_scoresvc)
roc_aucsvc = metrics.auc(fprsvc,tprsvc)
print('AUC of SVC Classifier:',roc_aucsvc)

y_scoredtr = dtr.predict_proba(X_train)[:,1]
fprdtr, tprdtr, threshdtr = metrics.roc_curve(y_train,y_scoredtr)
roc_aucdtr = metrics.auc(fprdtr,tprdtr)
print('AUC of DTR Classifier:',roc_aucdtr)

y_scoreann = ann.predict_proba(X_train)[:,1]
fprann, tprann, threshann = metrics.roc_curve(y_train,y_scoreann)
roc_aucann = metrics.auc(fprann,tprann)
print('AUC of ANN Classifier:',roc_aucann)

plt.figure(figsize=(6,6))
lw = 2
plt.plot(fprsvc, tprsvc, color='darkorange',
         lw=lw, label='SVC-ROC curve (area = %0.3f)' % roc_aucsvc)
plt.plot(fprdtr, tprdtr, color='darkred',
         lw=lw, label='DTR-ROC curve (area = %0.3f)' % roc_aucdtr)
plt.plot(fprann, tprann, color='green',
         lw=lw, label='ANN-ROC curve (area = %0.3f)' % roc_aucann)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")


# annotate thresholds at every other point
for i in range(1, len(threshsvc),2 ):
    plt.annotate('%0.2f' % threshsvc[i],xy=(fprsvc[i]+0.01,tprsvc[i]-0.03),
                 textcoords='data',fontsize=10,color='darkorange', rotation=-45  )
for i in range(1, len(threshdtr),1 ):
    plt.annotate('%0.2f' % threshdtr[i],xy=(fprdtr[i]+0.01,tprdtr[i]-0.03),
                 textcoords='data',fontsize=10,color='darkred', rotation=-45  )
for i in range(1, len(threshann),2 ):
    plt.annotate('%0.2f' % threshann[i],xy=(fprann[i]+0.01,tprann[i]-0.03),
                 textcoords='data',fontsize=10,color='green', rotation=-45  )


plt.show()
```

```python
#using Test Data Set

y_scoresvc1 = svc.predict_proba(X_test)[:,1]
fprsvc1, tprsvc1, threshsvc1 = metrics.roc_curve(y_test,y_scoresvc1)
roc_aucsvc1 = metrics.auc(fprsvc1,tprsvc1)
print('AUC of SVC Classifier:',roc_aucsvc1)

y_scoredtr1 = dtr.predict_proba(X_test)[:,1]
fprdtr1, tprdtr1, threshdtr1 = metrics.roc_curve(y_test,y_scoredtr1)
roc_aucdtr1 = metrics.auc(fprdtr1,tprdtr1)
print('AUC of DTR Classifier:',roc_aucdtr1)

y_scoreann1 = ann.predict_proba(X_test)[:,1]
fprann1, tprann1, threshann1 = metrics.roc_curve(y_test,y_scoreann1)
roc_aucann1 = metrics.auc(fprann1,tprann1)
print('AUC of ANN Classifier:',roc_aucann1)

plt.figure(figsize=(6,6))
lw = 2
plt.plot(fprsvc1, tprsvc1, color='darkorange',
         lw=lw, label='SVC-ROC curve (area = %0.3f)' % roc_aucsvc1)
plt.plot(fprdtr1, tprdtr1, color='darkred',
         lw=lw, label='DTR-ROC curve (area = %0.3f)' % roc_aucdtr1)
plt.plot(fprann1, tprann1, color='green',
         lw=lw, label='ANN-ROC curve (area = %0.3f)' % roc_aucann1)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")


# annotate thresholds at every other point
for i in range(1, len(threshsvc1),2 ):
    plt.annotate('%0.2f' % threshsvc1[i],xy=(fprsvc1[i]+0.01,tprsvc1[i]-0.03),
                 textcoords='data',fontsize=10,color='darkorange', rotation=-45  )
for i in range(1, len(threshdtr1),1 ):
    plt.annotate('%0.2f' % threshdtr1[i],xy=(fprdtr1[i]+0.01,tprdtr1[i]-0.03),
                 textcoords='data',fontsize=10,color='darkred', rotation=-45  )
for i in range(1, len(threshann1),2 ):
    plt.annotate('%0.2f' % threshann1[i],xy=(fprann1[i]+0.01,tprann1[i]-0.03),
                 textcoords='data',fontsize=10,color='green', rotation=-45  )


plt.show()
```

##Analysis & Discussion

1)
Based on the test data set, the overall best classifier is dependent on your weighting of false positives vs. true positives. There are many cases where it might not be optimal to choose a classifier that weights the two outcomes equally. If we were to choose them to be equal in weight, the SVC classifier yields the most optimal classifying mechanism. With a threshold value of .46 as the score split, the scores retreived by running the test data set through the classifier yields the highest TPR/FPR value. 

2) The Neural Network (ANN) proved to be the best classifier at achieving maximal TPR at the lowest FPR. The SVC also proved to have the highest TPR while at a minimal FPR. The one area where the DTR was optimal was that it obtained the highest AUC which is a good indicated of the overall scoring accuracy. 

3) Next I will use the SVC classifier to build a set of scores from the X_new dataset and the optimal threshold value (closest to ROC heaven)(.46) to determine how this classifier operates with this new dataset.


```python
#Testing X_new

for clf in clf_list:
    clf.fit(X_new,_)

#Define scores
_svc = svc.predict_proba(X_new)[:,1]  
_dtr = dtr.predict_proba(X_new)[:,1]
_ann = ann.predict_proba(X_new)[:,1]
#Define rates for ROC Curve
fprsvc2, tprsvc2, threshsvc2 = metrics.roc_curve(_,_svc)
fprdtr2, tprdtr2, threshdtr2 = metrics.roc_curve(_,_dtr)
fprann2, tprann2, threshann2 = metrics.roc_curve(_,_ann)
#Calculate Area Under Curve(AUC) for each Classifier
roc_aucsvc2 = metrics.auc(fprsvc2,tprsvc2)
roc_aucdtr2 = metrics.auc(fprdtr2,tprdtr2)
roc_aucann2 = metrics.auc(fprann2,tprann2)
#I tried printing out the AUC for each classifier but it did not yield much meaningful data
print('Actual Dataset: ',_)
print('SVC Dataset: ',_svc)
print('DTR Dataset: ',_dtr)
print('ANN Dataset: ',_ann)
print("SVC with .46 threshold:")
print("predicted - actual")
for ind, score in enumerate(_svc):
    if score > 0.46:
        if _[ind] == 1:
            print("  1         " + str(_[ind]))
        else:
            print("  1         " + str(_[ind]))
    else:
        print("  0         " + str(_[ind]))

print("DTR with .96 threshold:")
print("predicted - actual")
for ind, score in enumerate(_dtr):
    if score > 0.96:
        if _[ind] == 1:
            print("  1         " + str(_[ind]))
        else:
            print("  1         " + str(_[ind]))
    else:
        print("  0         " + str(_[ind]))
        
print("ANN with .34 threshold:")
print("predicted - actual")
for ind, score in enumerate(_ann):
    if score > 0.34:
        if _[ind] == 1:
            print("  1         " + str(_[ind]))
        else:
            print("  1         " + str(_[ind]))
    else:
        print("  0         " + str(_[ind]))
```

   Using the most optimal threshold values from the previous dataset for each classifier, the new data set presents quite different results for the much small data set. The .46 threshold value for the SVC classifier resulted in a 1.0 TPR however a 1.0 FPR which is not ideal and is no different than flat out claiming that all belong to the 'true' class. Both the DTR and the ANN produced a 1.0 TPR and a 0.0 FPR which puts them both, right smack dab in the middle of of 'ROC heaven.'
