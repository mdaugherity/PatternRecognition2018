# Neural Network Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import neural_network
```

```python
X = np.linspace(0,9,10)  
X2d = X.reshape(-1,1) # need a 2D array (n_samples, n_features)
```

```python
# Start simple, linear + noise
yLinear = 2*X - 3 + 3*np.random.rand(X.size)
plt.scatter(X,yLinear)
plt.show()
```

```python
mlp = neural_network.MLPRegressor(hidden_layer_sizes=(4), max_iter=5000)
mlp.fit(X2d,yLinear)
```

```python
Xpred = np.linspace(0,9,100).reshape(-1,1)  # remember X is a 2D array (n_samples, n_features)
```

```python
ypred = mlp.predict(Xpred)
```

```python
plt.scatter(X,yLinear)
plt.plot(Xpred,ypred,'r-')
plt.show()
```

---
## now test multiple hidden layer configurations

```python
hiddens = ((2),(4),(6),(10),(50),(100),(5,5),(10,10))  # list of hidden layers/nodes to try

plt.figure(figsize=(16,9))

for i,hidden in enumerate(hiddens):
    ax = plt.subplot(2,4,i+1)
    ax.set_title(hidden)
    mlp = neural_network.MLPRegressor(hidden_layer_sizes=hidden,max_iter=10000,random_state=42)
    mlp.fit(X2d,yLinear)
    ypred = mlp.predict(Xpred)
 
    plt.scatter(X,yLinear)
    plt.plot(Xpred,ypred,'r-')
plt.show()
```

```python
# Try non-linear
yCos = np.sin(0.5*X)
plt.scatter(X,yCos)
plt.show()
```

```python
hiddens = ((2),(4),(6),(10),(50),(100),(5,5),(100,100))  # list of hidden layers/nodes to try
hiddens = ((5),(10),(50),(100),(5,5),(10,10),(50,50),(100,100))  # list of hidden layers/nodes to try

plt.figure(figsize=(16,9))

for i,hidden in enumerate(hiddens):
    ax = plt.subplot(2,4,i+1)
    ax.set_title(hidden)
    mlp = neural_network.MLPRegressor(hidden_layer_sizes=hidden,max_iter=10000,random_state=41)
    mlp.fit(X2d,yCos)
    ypred = mlp.predict(Xpred)
 
    plt.scatter(X,yCos)
    plt.plot(Xpred,ypred,'r-')
plt.show()
```
