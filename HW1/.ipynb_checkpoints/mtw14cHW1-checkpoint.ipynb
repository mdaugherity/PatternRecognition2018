## Problem Description
Given three classifiers, determine which is best "overall", which has the best precision, and which minimizes false positives the most. Finally, I will find the optimal probability threshold and use it to classify the points in X_new.


## Solution Method
First I will build the data sets and classifiers as demonstrated in the description, then I will plot all three as ROC curves and hopefully that will provide some insight as to the relative accuracies of each.


## Input
Load and pre-process data

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# first we make a data set to train and test on
X, y = datasets.make_classification(n_samples=100, n_features=8, n_informative=3, 
                                      n_classes=2, class_sep=0.7, random_state=1)

# we divide this into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# then we create a new dataset to classify later with our best classifier
X_new, _ = datasets.make_classification(n_samples=5, n_features=8, n_informative=3, 
                                      n_classes=2, class_sep=0.7, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn import tree
from sklearn import neural_network

# we create three classifiers, an SVC, a decision tree, and a neural network
svc = SVC(kernel='poly', C=1,probability=True)
dtr = tree.DecisionTreeClassifier(max_depth=4)
ann = neural_network.MLPClassifier(random_state=1)

# we combine them into a list for ease of use
clf_list = (svc,dtr,ann) # list o' classifiers

# Train all the classifiers in a for loop, just because we can
for clf in clf_list:
    clf.fit(X_train,y_train)
    
# Expect to see an error message about the neural network not converging below...
```

## Analysis

Here we plot the ROC graphs for each of the three classifiers. First I made the ROC curves using the training data set. You will notice that they are extrmely accurate, which led me to realize that I should be using the test data instead, which follows. 

```python
import matplotlib.pyplot as plt
from sklearn import metrics
#calculate scores for each classifier
y_score1 = svc.predict_proba(X_train)[:,1]
y_score2 = dtr.predict_proba(X_train)[:,1]
y_score3 = ann.predict_proba(X_train)[:,1]
# calculate roc_curve for each classifier
fpr1, tpr1, thresh1 = metrics.roc_curve(y_train,y_score1)
fpr2, tpr2, thresh2 = metrics.roc_curve(y_train,y_score2)
fpr3, tpr3, thresh3 = metrics.roc_curve(y_train,y_score3)

# set up roc_curve plot 
plt.figure(figsize=(6,6))
lw = 2

# add ROC curves to the plot
plt.plot(fpr1, tpr1, color='darkorange',
         lw=lw, label='SVC ROC curve (area = %0.3f)' % metrics.auc(fpr1,tpr1))
plt.plot(fpr2, tpr2, color='darkgreen',
         lw=lw, label='Tree ROC curve (area = %0.3f)' % metrics.auc(fpr2,tpr2))
plt.plot(fpr3, tpr3, color='darkred',
         lw=lw, label='Neural Net. ROC curve (area = %0.3f)' % metrics.auc(fpr3,tpr3))
# more setup
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Training Data Set (WRONG)')
plt.legend(loc="lower right")

# annotate thresholds at every other point
for i in range(1, len(thresh1),2 ):
    plt.annotate('%0.2f' % thresh1[i],xy=(fpr1[i]+0.01,tpr1[i]-0.03),
                 textcoords='data',fontsize=10,color='darkorange', rotation=-45  )
for i in range(1, len(thresh2) ):
    plt.annotate('%0.2f' % thresh2[i],xy=(fpr2[i]+0.01,tpr2[i]-0.03),
                 textcoords='data',fontsize=10,color='darkgreen', rotation=-45  )
for i in range(1, len(thresh3),2 ):
    plt.annotate('%0.2f' % thresh3[i],xy=(fpr3[i]+0.01,tpr3[i]-0.03),
                 textcoords='data',fontsize=10,color='darkred', rotation=-45  )

plt.show()
```

```python

#calculate scores for each classifier
y_score1 = svc.predict_proba(X_test)[:,1]
y_score2 = dtr.predict_proba(X_test)[:,1]
y_score3 = ann.predict_proba(X_test)[:,1]
# calculate roc_curve for each classifier
fpr1, tpr1, thresh1 = metrics.roc_curve(y_test,y_score1)
fpr2, tpr2, thresh2 = metrics.roc_curve(y_test,y_score2)
fpr3, tpr3, thresh3 = metrics.roc_curve(y_test,y_score3)

# set up roc_curve plot 
plt.figure(figsize=(6,6))
lw = 2

# add ROC curves to the plot
plt.plot(fpr1, tpr1, color='darkorange',
         lw=lw, label='SVC ROC curve (area = %0.3f)' % metrics.auc(fpr1,tpr1))
plt.plot(fpr2, tpr2, color='darkgreen',
         lw=lw, label='Tree ROC curve (area = %0.3f)' % metrics.auc(fpr2,tpr2))
plt.plot(fpr3, tpr3, color='darkred',
         lw=lw, label='Neural Net. ROC curve (area = %0.3f)' % metrics.auc(fpr3,tpr3))
# more setup
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Testing Data Set')
plt.legend(loc="lower right")

# annotate thresholds at every other point
for i in range(1, len(thresh1),2 ):
    plt.annotate('%0.2f' % thresh1[i],xy=(fpr1[i]+0.01,tpr1[i]-0.03),
                 textcoords='data',fontsize=10,color='darkorange', rotation=-45  )
for i in range(1, len(thresh2) ):
    plt.annotate('%0.2f' % thresh2[i],xy=(fpr2[i]+0.01,tpr2[i]-0.03),
                 textcoords='data',fontsize=10,color='darkgreen', rotation=-45  )
for i in range(1, len(thresh3),2 ):
    plt.annotate('%0.2f' % thresh3[i],xy=(fpr3[i]+0.01,tpr3[i]-0.03),
                 textcoords='data',fontsize=10,color='darkred', rotation=-45  )

plt.show()
```
